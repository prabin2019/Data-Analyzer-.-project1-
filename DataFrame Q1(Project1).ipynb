```python
#To plot a graph of a certain data first of all we need to find out the missing standard  non standard values  and replace it with appropriate values .
```

```python
MAC270['INAUGURATION_AGE'].isnull()
```

```python
median = MAC270['INAUGURATION_AGE'].median()
MAC270['INAUGURATION_AGE'].fillna(median,inplace=True)

```

```python
MAC270
```

```python
plt.figure(figsize=(13,6))
plt.xlabel('NAME')
plt.ylabel('INAUGURATION_AGE')
plt.title('The Inauguration Age of U.S Presidents')
plt.plot(MAC270['NAME'], MAC270['INAUGURATION_AGE'])
```
